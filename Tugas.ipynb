```python
path = '/content/drive/MyDrive/subaru duck.jpg'
```

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
image = cv.imread(path)
cv2_imshow(image)
```

```python
image2 = image.copy()
image2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
cv2_imshow(image2)
```

```python
tinggi = image2.shape[0]
lebar = image2.shape[1]
print("resolusi image : tinggi x lebar = ", tinggi, "x ", lebar)

image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (lebar):
  image2[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image, image2))
cv2_imshow(final_frame)
```

```python
image2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (lebar):
  image3[int(y-155),int(lebar-105)] = [255,255,255]
for y in range (tinggi):
  image4[y,int(y*(lebar/tinggi))] = [255,255,255]
  image4[y,lebar-1-int(y*(lebar/tinggi))] = [255,255,255]

final_frame = cv.hconcat((image2, image3))
final_frame2 = cv.hconcat((image2, image4))
cv2_imshow(final_frame)
print('\n')
cv2_imshow(final_frame2)
```

```python
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (100):
  image2[int((tinggi)/2),y+120] = [255,255,255]

final_frame = cv.hconcat((image, image2))
cv2_imshow(final_frame)
```

```python
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image 
for y in range(90,175): 
  for x in range (105, 175):
    image2[x, y] = [255,255,255]


final_frame = cv.hconcat ((image, image2))
cv2_imshow(final_frame)
```
